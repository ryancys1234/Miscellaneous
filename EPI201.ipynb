```python
import matplotlib.pyplot as plt, numpy as np

x_1 = [0, 3, 9, 19, 24]
x_2 = [0, 3, 5, 7, 24]
y_1 = [100, 100, 75, 37.5, 18.75]
y_2 = [100, 100, 62.5, 37.5, 12.5]

plt.step(x_1, y_1, label = 'Treatment arm'); plt.step(x_2, y_2, label = 'Placebo arm')
for i in range(3): plt.annotate(f'{y_1[i+2]}%', (x_1[i + 1], y_1[i + 2]), textcoords = 'offset points', xytext = (5, 7))
for i in range(3): plt.annotate(f'{y_2[i+2]}%', (x_2[i + 1], y_2[i + 2]), textcoords = 'offset points', xytext = (5, 7))
plt.title('Kaplan-Meier curves for cumulative survival vs the follow-up time')
plt.xlabel('Time (months)'); plt.ylabel('Cumulative survival (%)')
plt.xticks(np.arange(0, 25, 2)); plt.yticks(np.arange(0, 101, 20))
plt.legend()
plt.show()
```
